#Custom TOC builder for Ecofoci_CodeLab 

- specifically the section for field data

```python
import glob
import pandas as pd
```

```python
df = pd.DataFrame([x.split('/') for x in sorted(glob.glob('../*[0-9]/*/*ipynb'))],columns=['dir','year','mooring','files'])
```

```python
df
```

```python
codebook_path = 'content/code_gallery/ecofoci_fielddata_history/'

for year,ydata in df.groupby('year'):
    print(f'- file: {codebook_path}{year}/index.md')
    print('   sections:')
    for mooring,mdata in ydata.groupby('mooring'):
        print(f'     - file: {codebook_path}{year}/{mooring}/index.md')
        print('      sections:')
        for files,fdata in mdata.groupby('files'):
            print(f'      - file: {codebook_path}{year}/{mooring}/{files}')
            print(f'        title: {files.split(".")[0]}')
```
