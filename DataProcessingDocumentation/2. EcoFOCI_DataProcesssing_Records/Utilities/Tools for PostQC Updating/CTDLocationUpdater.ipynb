```python
import xarray as xa
import yaml
import os
```

```python
sample_data_dir = '/Users/bell/ecoraid/2022/CTDcasts/dy2208l2/'
```

```python
###############################################################
# edit to point to {cruise sepcific} raw datafiles 
datafile = sample_data_dir+'final_data_cf/ctd/' #<- point to cruise and process all files within
cruise_name = 'DY2208L2' #no hyphens
cruise_meta_file = sample_data_dir+'logs/DY2208L2.yaml'
inst_meta_file = sample_data_dir+'logs/FOCI_standard_CTD.yaml'
# group_meta_file = ecofocipy_dir+'staticdata/institutional_meta_example_ctd.yaml'
inst_shortname = '' #keep as placeholder for now
###############################################################
```

```python
#cruise info
with open(cruise_meta_file) as file:
    cruise_config = yaml.full_load(file)
```

```python
cruise_config
```

```python
for filename in os.listdir(datafile):
    if 'nc' in filename:
        print(filename)
        xdf = xa.load_dataset(datafile+filename)

        longitude = float(cruise_config['CTDCasts'][f"CTD{filename.split('c')[1].split('_')[0]}"]['LongitudeDeg']) + \
            (float(cruise_config['CTDCasts'][f"CTD{filename.split('c')[1].split('_')[0]}"]['LongitudeMin']))/60
        latitude = float(cruise_config['CTDCasts'][f"CTD{filename.split('c')[1].split('_')[0]}"]['LatitudeDeg']) + \
            (float(cruise_config['CTDCasts'][f"CTD{filename.split('c')[1].split('_')[0]}"]['LatitudeMin']))/60
        
        longitude = -1*longitude
        xdf['latitude'] = [latitude]
        xdf['longitude'] = [longitude]
        
        xdf.to_netcdf(datafile+filename,format="NETCDF3_CLASSIC",encoding={'time':{'units':'days since 1900-01-01'}})
```
