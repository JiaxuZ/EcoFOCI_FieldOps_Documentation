# TimeSeries Mooring Data - Quick Plot

From an erddap Dataset, specify the Mooring ID to get a quick plot of the data

```python
from erddapy import ERDDAP
import pandas as pd
import numpy as np
```

```python
server_url = 'http://ecofoci-field.pmel.noaa.gov:8080/erddap'
dataset_name = 'ADCP_Mooring_22upp3a_final'
e = ERDDAP(server=server_url,
          protocol='griddap',
          response='ncCF')

e.dataset_id=dataset_name

pdf = e.to_xarray()
pdf
```

```python
for i,groups in pdf.groupby('depth'):
    print(i)
```

```python
with pd.ExcelWriter(f'{dataset_name}.xlsx') as writer:  

    for i,groups in pdf.groupby('depth'):
        groups.to_dataframe().to_excel(writer, sheet_name=f'{i}_m')
```
