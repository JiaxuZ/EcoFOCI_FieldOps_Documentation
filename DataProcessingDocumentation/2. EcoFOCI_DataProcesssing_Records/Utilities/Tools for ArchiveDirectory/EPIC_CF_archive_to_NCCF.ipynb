# EPIC Archive (with CF time) to NCCF Archive

In the event data needs to be reprocessed from an old completed mooring or cruise, it would be best to convert all the files to the 2020+ CF convention.

These means instead of *.cf.nc files converted from old EPIC files, we will make *.nc files in the final_data_cf folder from the *.cf.nc files in the variable names only:
- lat -> latitude
- lon -> longitude
  

```python
epic2cf_fluor = {
    'lat':'latitude',
    'lon':'longitude',
    'fluor_3031':'fluorometer_volts',
    'Fch_906':'chlor_fluorescence',
}
    
epic2cf_seacat = {
    'lat':'latitude',
    'lon':'longitude',
    'S_41':'salinity',
    'T_20':'temperature',
    'P_1':'pressure',
    'C_50':'conductivity',
}

epic2cf_s37 = {
    'lat':'latitude',
    'lon':'longitude',    
    'S_41':'salinity',
    'T_20':'temperature',
    'P_1':'pressure',
}

epic2cf_s39 = {
    'lat':'latitude',
    'lon':'longitude',    
    'T_20':'temperature',
    'P_1':'pressure',
}
```

```python
import xarray as xa
```

```python
path = '/Users/bell/ecoraid/2017/Moorings/17ckp4a/erddap/final_data/'
name = '17ckp4a_sc_0039m.cf.nc'

xdf = xa.load_dataset(f'{path}{name}')
xdf
```

```python
xdf = xdf.rename(epic2cf_seacat)
xdf
```

```python
xdf['latitude'] = xdf['latitude'] * 0 + 62.194883
xdf['longitude'] = xdf['longitude'] * 0 + -174.683483
# xdf['depth'] = xdf['depth'] * 0 + 31
```

```python
xdf.to_netcdf(path=f'{name.split('.')[0]}.nc',format="NETCDF3_CLASSIC",
              encoding={'time':{'units':'days since 1900-01-01'}})
```
